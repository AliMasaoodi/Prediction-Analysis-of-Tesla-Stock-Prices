```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas
import numexpr
```

# Initial analysis

## Data Importing

```python
tesla = pd.read_csv('C:/Users/Mohammed/Downloads/TSLAStock.csv')
tesla.head()
```

## DataFrame statistics

```python
tesla.set_index('Date')
```

## Stock data line chart

```python
plt.figure(figsize=(15, 7))
tesla['Close'].plot()
plt.show()
```

## Price difference

```python
tesla['PriceDiff'] = tesla['Close'].shift(-1) - tesla['Close']
print(tesla['PriceDiff'])
```

## Daily Stock return

```python
tesla['Return'] = tesla['PriceDiff'] /tesla['Close']
print(tesla['Return'])
```

# Rolling Window Analysis 

```python
tesla['ma50'] = tesla['Close'].rolling(50).mean()
#plot the moving average
plt.figure(figsize=(15, 8))
tesla['ma50'].plot(label='MA50')
tesla['Close'].plot(label='Close')
plt.legend()
plt.show()
```
